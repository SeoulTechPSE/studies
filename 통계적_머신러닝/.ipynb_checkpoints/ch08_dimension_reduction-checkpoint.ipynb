# 8장. 차원축소(dimension reduction)

\begin{align*}
 S^2 &= \frac{1}{n} \sum_{i=1}^n \mathbf{x}_i \mathbf{x}_i^T = \frac{1}{n}\mathbf{X}^T \mathbf{X}, \;\;\mathbf{x}_i \in \mathbb{R}^{d\times 1}\\ 
 &\big\Downarrow\;\;\text{가장 큰 }l\text{개의 고윳값에 대응하는 } l\text{개의 고유백터로 구성된 행렬 } \mathbf{W}\\ 
 \mathbf{z}_i &=\mathbf{W}^T \mathbf{x}_i\\ 
\end{align*}

## 8.1 특이값 분해(Singular Value Decomposition, SVD)

\begin{align*}
 \mathbf{X}_{n\times d}&= \mathbf{U}_{n\times n} \boldsymbol{\Lambda}_{n\times d} \mathbf{V}_{d\times d}^T \\ \mathbf{U}^T &\mathbf{U}=\mathbf{I},\;\;\mathbf{V}^T \mathbf{V}=\mathbf{I}, \;\;\mathbf{V}^T \mathbf{V}=\mathbf{I}, \;\;r=\min(n,d) \\ 
 &\Downarrow\;\; \text{when }n \geq d  \\ 
\mathbf{X}_{n\times d}&= \tilde{\mathbf{U}}_{n\times d} \tilde{\boldsymbol{\Lambda}}_{d\times d} \mathbf{V}_{d\times d}^T  \\
 &\Downarrow \\
 \mathbf{X}^T\mathbf{X} &=\mathbf{V}\tilde{\boldsymbol{\Lambda}} \tilde{\mathbf{U}}^T  \tilde{\mathbf{U}} \tilde{\boldsymbol{\Lambda}} \tilde{\mathbf{V}}^T \\
 &\Downarrow \\
 (\mathbf{X}^T\mathbf{X}) \mathbf{V} &=  \mathbf{V} \tilde{\boldsymbol{\Lambda}}^2 \\
 &\Downarrow \\
  \mathbf{V}\text{는 고유백터 행렬}, \;\; \tilde{\boldsymbol{\Lambda}}^2\text{은 고유값}, 
  \;\;&\text{ PCA는 가장 큰 } l\text{개의 고윳값만 사용} \\ 
 &\Downarrow \\  
 \mathbf{X}_{n\times d}&\approx  \mathbf{U}_{n\times l}^* \boldsymbol{\Lambda}_{l\times l}^* \mathbf{V}_{l\times d}^{*T} 
\end{align*}
