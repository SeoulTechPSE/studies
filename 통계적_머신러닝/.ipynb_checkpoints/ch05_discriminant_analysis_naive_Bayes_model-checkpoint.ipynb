# 5장 판별분석과 단순 베이즈 모형

$$P(y=k | \mathbf{x}) = \frac{P(\mathbf{x} | k) P(k)}{P(\mathbf{x})} = \frac{P(\mathbf{x} | k) P(k)}{\displaystyle \sum_{k=1}^K P(\mathbf{x} | k) P(k)}$$

## 5.1 판별분석(discriminant analysis)

\begin{align*}
 P(\mathbf{x} | k) &= \frac{1}{(2\pi)^{1/2}|\Sigma_k|^{1/2}} 
  \exp \left[ -\frac{1}{2} (\mathbf{x} -\boldsymbol{\mu}_k)^T \Sigma_k^{-1} (\mathbf{x} -\boldsymbol{\mu}_k) \right]\\ 
 &\Big\Downarrow \;\;\Sigma_k \sim \Sigma \\ 
 \log \frac{P(y=k | \mathbf{x})}{P(y=l | \mathbf{x})}&= \log \frac{P(\mathbf{x} | k) P(y=k)}{P(\mathbf{x} | l) P(y=l)} \\ 
 &= \log \frac{P(y=k)}{P(y=l)} 
   -\frac{1}{2} (\boldsymbol{\mu}_k +\boldsymbol{\mu}_l)^T \Sigma^{-1} (\boldsymbol{\mu}_k -\boldsymbol{\mu}_l)
   +\mathbf{x}^{T} \Sigma^{-1} (\boldsymbol{\mu}_k -\boldsymbol{\mu}_l) \\
  &\Big\Downarrow \;\; \delta_k(\mathbf{x}) =\mathbf{x}^{T} \Sigma^{-1} \boldsymbol{\mu}_k
   -\frac{1}{2} \boldsymbol{\mu}_k^T \Sigma^{-1} \boldsymbol{\mu}_k + \log \pi_k \;\;
    \text{ : linear discriminant function}\\
 \log \frac{P(y=k | \mathbf{x})}{P(y=l | \mathbf{x})}&= \delta_k(\mathbf{x}) - \delta_l(\mathbf{x})
\end{align*}

\begin{align*}
 \hat{\pi}_k &= \frac{n_k}{n} \\ 
 \hat{\boldsymbol{\mu}_k} &= \frac{1}{n_k}\sum_{i=1}^{n_k} \mathbf{x}_i^{(k)} \\ 
 \hat{\Sigma} &= \frac{1}{n-K}\sum_{k=1}^K \sum_{i=1}^{n_k} \left( \mathbf{x}_i^{(k)} -\hat{\boldsymbol{\mu}}_k \right)
  \left( \mathbf{x}_i^{(k)} -\hat{\boldsymbol{\mu}}_k \right ) ^T  
\end{align*}

* 이차 판별함수 (quadratic discriminant function)


\begin{align*}
 \delta_k(\mathbf{x}) &= \log P(\mathbf{x} | k ) P(y=k)\\
  &\Big\Downarrow \;\; \Sigma_k \neq \Sigma \\
  \delta_k(\mathbf{x}) &=-\frac{1}{2} \log |\Sigma_k| 
    -\frac{1}{2} (\mathbf{x} -\boldsymbol{\mu}_k)^T \Sigma_k^{-1} (\mathbf{x} -\boldsymbol{\mu}_k) + \log \pi_k \;\;
    \text{ : quadratic discriminant function}
\end{align*}

## 5.3. Scikit learn을 이용한 LDA와 단순 베이즈 모형

```python
import seaborn as sns

iris = sns.load_dataset('iris')

X = iris.drop('species', axis=1)
y = iris['species']
```

```python
from sklearn.preprocessing import LabelEncoder

classle = LabelEncoder()
y = classle.fit_transform(iris['species'].values)
```

---

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123, stratify=y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

cld = LinearDiscriminantAnalysis(store_covariance=True)
cld.fit(X_train, y_train)

y_train_pred = cld.predict(X_train)
y_test_pred = cld.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

```python
print(cld.means_)
```

```python
print(cld.xbar_)
```

```python
print(cld.covariance_)
```

```python
print(cld.intercept_)
```

```python
print(cld.coef_)
```

---

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_pred))
```

```python
print(cld.predict_proba(X_test[:5]))
```

---

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

cqd = QuadraticDiscriminantAnalysis()
cqd.fit(X_train, y_train)

y_train_pred = cqd.predict(X_train)
y_test_pred = cqd.predict(X_test)
```

```python
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

```python
print(confusion_matrix(y_test, y_test_pred))
```

```python
from sklearn.datasets import load_digits

digits = load_digits()
print(digits.images.shape)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(18, 8), subplot_kw={'xticks': [], 'yticks': []},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]), transform=ax.transAxes, color='red')
```

```python
X = digits.data
y = digits.target

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_test_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(y_test, y_test_pred)
print(mat)
```
