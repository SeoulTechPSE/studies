## 6.1 회귀나무(Regression Tree)

1. 특성변수 각각의 관측치를 오름차순으로 정렬한다. 예를 들어 $x_2$의 관측치가 $\{ -5, 1, -3, 3 \}$이라면 $\{ -5, -3, 1, 3 \}$으로 정리한다.


2. 정렬된 $x_1$의 관측치를 차례대로 노드로 하여 분할한 후, 두 개의 셀(왼쪽과 오른쪽이라는 의미에서 이를 $l$, $r$로 표현한다) 각각에 포함된 평균 $\bar{y}_t^l$, $\bar{y}_t^r$을 구한다. 그리고

   $$\text{SSE}(x_1)= \sum_{i \in l} \left( y_i - \bar{y}_t^l\right)^2 +\sum_{i \in r} \left( y_i - \bar{y}_t^r\right)^2 $$
   
   를 최소화하는 $x_1$의 관측치를 $x_1$의 노드로 선택한다. 예를 들어 $x_1=\{ -1, 0, 2, 3\}$이면 4개의 $\text{SSE}(x_1)$가 계산되고, 이 중 $x_1=0$이 가장 작은 $\text{SSE}(x_1)$를 가지고 있다면 $x_1$의 노드는 0이 된다. 이러한 과정을 $x_1, x_2, \cdots, x_d$에 대해 실시하여 각각 최소의 $\text{SSE}(x_1)$, $\text{SSE}(x_2)$, $\cdots$, $\text{SSE}(x_d)$을 갖는 노드를 구한다.


3. 2단계에서 $d$개의 최소 $\text{SSE}(x_1)$, $\text{SSE}(x_2)$, $\cdots$, $\text{SSE}(x_d)$ 중 가장 작은 최소 $\text{SSE}$를 가진 특징변수를 $x^*$라고 표현하고, $x^*$의 노드를 기준으로 첫번 째 분할을 실시하여 2개의 영역을 만든다.


4. 3단계에서 구한 2개의 영역 각각에 대해 영역 내에서 1-3단계를 반복한다. 그리고 결과로 나온 영역들에 대해서도 이를 반복한다.

* 새로운 관측치 $\mathbf{x}$에 대한 회귀나무의 예측치


$$\hat{y} = \sum_{k=1}^m \bar{y}_k I(\mathbf{x} \in R_k)$$

* 의사 결정나무는 각 셀에 오직 한 개의 학습데이타만 존재할 때까지 반복할 수 있으므로 과대 적합 문제가 발생할 수 밖에 없다. 그러므로 분할의 깊이를 제안할 필요가 있으며, 이를 가지치기(pruning)라고 한다. 

$$C_\alpha (T) = \sum_{m=1}^T \text{SSE}_m +\alpha T, \;\;T\text{: 총 셀의 갯수}$$

## 6.2 분류나무(Classification Tree)

* $\hat{P}_{mk}$를 셀 $R_m$에 있는 범주 $k$의 비율이라고 정의할 때, 분류나무에서는 다음의 불순도(impurity) 측도들을 사용한다:

\begin{align*}
 \text{Gini Index}&: I_G(R_m)=\sum_{k=1}^K \hat{P}_{mk} \left(1-\hat{P}_{mk} \right)=1-\sum_{k=1}^K \hat{P}_{mk}^2\\ 
 \text{Cross Entropy}&: I_C(R_m) =-\sum_{k=1}^K \hat{P}_{mk} \log_2 \hat{P}_{mk}\\ 
 \text{Misclassification Error}&: I_E(R_m)=1 -\max_k \hat{P}_{mk}
\end{align*}

* 상위 셀 $R$을 $x_j$를 기준으로 두 개의 영역 $R_l$, $R_r$로 나누어 불순도 측도 $I(R)$, $I(R_l)$, $I(R_r)$를 계산할 수 있다. $N_R$, $N_{R_l}$, $N_{R_r}$을 각각 $R$, $R_l$, $R_r$에 포함된 학습데이터의 개수라고 할 때, 정보획득(Information Gain, IG)은 다음과 같다:

$$IG(R, x_j) = I(R) -\frac{N_{R_l}}{N_{R}} I(R_l) -\frac{N_{R_r}}{N_R} I(R_r)$$

* 회귀나무의 2, 3단계에서 사용한 $\text{SSE}(x_j)$ 대신 $IG(R, x_j)$를 이용하되 $IG(R, x_j)$의 최댓값을 사용해 분할 변수와 노드를 선택한다

```python
import numpy as np
import matplotlib.pyplot as plt

def gini(p):
    return p*(1-p) + (1 - p)*(1 - (1 - p))

def entropy(p):
    eps = 1.0e-14
    return -p*np.log2(p +eps) - (1 - p)*np.log2(1 - p +eps)

def error(p):
    return 1 - np.max([p, 1 - p], axis=0)
```

```python
p = np.linspace(0.0, 1.0)

fig, ax = plt.subplots()

for impurity, label, ls, color, in zip([entropy(p), 0.5*entropy(p), gini(p), error(p)],
                                       ['Entropy', 'Entropy(scaled)', 'Gini impurity', 'classification errot'],
                                       ['-', '-', '--', '-.'],
                                       ['black', 'lightgray', 'red', 'green']):
    ax.plot(p, impurity, label=label, ls=ls, lw=2, color=color)
    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), frameon=False, ncol=4)
ax.axhline(y=0.5, lw=1, color='k', ls=':')
ax.axhline(y=1.0, lw=1, color='k', ls=':')
plt.ylim([0, 1.1])
plt.xlabel('$p_1$')
plt.ylabel('impurity')
plt.show()
```

## 6.3 Scikit learn을 이용한 의사결정나무

```python
import seaborn as sns

iris = sns.load_dataset('iris')

X = iris.drop('species', axis=1)
y = iris['species']
```

```python
from sklearn.preprocessing import LabelEncoder

classle = LabelEncoder()
y = classle.fit_transform(iris['species'].values)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123, stratify=y)
```

---

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=1)
dtc.fit(X_train, y_train)

y_train_pred = dtc.predict(X_train)
y_test_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

```python
print(confusion_matrix(y_test, y_test_pred))
```

```python
from sklearn.tree import plot_tree

plot_tree(dtc.fit(X_train, y_train))
plt.show()
```

```python
# conda install -c conda-forge pydotplus

from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

dot_data = export_graphviz(dtc,
                           filled=True, 
                           rounded=True,
                           class_names=['Setosa', 'Versicolor', 'Virginica'],
                           feature_names=['sepal length', 'sepal width', 'petal length', 'petal width'],
                           out_file=None) 

graph = graph_from_dot_data(dot_data)
graph.write_png('./figs/tree_model.png')
Image('./figs/tree_model.png', width=600)
```
